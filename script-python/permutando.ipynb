```python
#LISTA DE PALAVRAS QUE COMEAM COM M, permutando onoMpled

import itertools

# Definindo a palavra
palavra = "onoMpedl"

# Removendo a letra M da palavra
outras_letras = palavra.replace('M', '')

# Obtendo todas as permutaes das letras, exceto a primeira
permutacoes = itertools.permutations(outras_letras)

# Construindo as palavras permutadas com "M" como primeira letra
palavras_permutadas = ['M' + ''.join(permutacao) for permutacao in permutacoes]

# Imprimindo as palavras permutadas
for palavra_permutada in palavras_permutadas:
    print(palavra_permutada)

```
